```python
import pandas as pd

import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
```

# Environment Datasets

### Loading the dataset constructed as presented in the notebook WeatherTideDigitalTwinDatasetCreation
### combining data from https://power.larc.nasa.gov/data-access-viewer/ and https://www.bodc.ac.uk/data/hosted_data_systems/sea_level/uk_tide_gauge_network/processed/#MyRequest

```python
data_dr = "./data/"
```

```python
ds_env = pd.read_csv(data_dr+"WeatherTide_digitaltwin_1h.csv")
```

# Pollution Dataset

## Loading the dataset constructed in PollutionDigitalTwinDatasetCreation

```python
df_poll = pd.read_csv(data_dr+"Pollution_1h.csv")
```

## Combining the environment and pollution digital twin for the Portsmouth harbour

```python
ds_env.head()
```

```python
df_poll.head()
```

# join environment and pollution twins

```python
df_e_p = pd.merge(ds_env, df_poll, on="date", how="outer")

df_e_p = df_e_p.sort_values("date").reset_index(drop=True)
```

```python
missing = df_e_p.isnull().sum()
missing = missing[missing > 0]
missing.sort_values(inplace=True)
f, ax = plt.subplots(figsize=(15, 7))


missing.plot.bar(label="number of NaNs");
plt.axhline(y=df_e_p.shape[0], color="red", label="number of observations")
plt.legend(fontsize=16)
plt.show()
```

```python
ds_poll = ds_poll[(ds_poll.date.dt.year<=2023)]
```

```python
ds_twin_ep = ds_env.join(ds_poll.drop("date", axis=1), how='outer')
```

```python
ds_twin_ep
```

## Studying the NaNs in the Dataset


```python
missing = ds_twin_ep.isnull().sum()
missing = missing[missing > 0]
missing.sort_values(inplace=True)
f, ax = plt.subplots()


missing.plot.bar(label="number of NaNs");
plt.axhline(y=ds_twin_ep.shape[0], color="red", label="number of observations")
plt.legend(fontsize=16)
plt.show()
```

# investigate the correlations

```python
corrmat = ds_twin_ep.drop("date", axis=1).astype(float).corr().dropna(how="all", axis="columns").dropna(how="all", axis="rows")
f, ax = plt.subplots( figsize=(15, 15))
fig = sns.heatmap(corrmat, vmax=1, square=True, cmap="RdBu",  annot=True, vmin=-1);

plt.show()
```

Let's create a subset with most important quantities

```python
set(ds_twin_ep.columns)
```

```python
subset_columns = [
 'Carbon monoxide',
 'Nitric oxide',
 'Nitrogen dioxide',
 'Nitrogen oxides as nitrogen dioxide',  'Ozone', 'PRECTOTCORR',
 'PS',
 'QV2M',
 'Sulphur dioxide', 'T2M', 'WD10M',
 'WD50M',
 'WS10M',
 'WS50M','value',

]
```

```python
corrmat = ds_twin_ep[subset_columns].astype(float).corr().dropna(how="all", axis="columns").dropna(how="all", axis="rows")
f, ax = plt.subplots( figsize=(15, 15))
fig = sns.heatmap(corrmat, vmax=1, square=True, cmap="RdBu",  annot=True, vmin=-1);

plt.show()
```
