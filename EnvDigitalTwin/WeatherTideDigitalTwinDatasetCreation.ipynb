# Height of tide water 

We make use of data from here 
https://www.bodc.ac.uk/data/hosted_data_systems/sea_level/uk_tide_gauge_network/processed/#MyRequest

## Let's construct the dataset with 15 minutes time resolution 2000-2023


```python
pd.read_csv(env_data_dr+'/BODC_tidal/2002PTM.txt',\
            skiprows=range(0,11), delim_whitespace=True, header=None,names=['index', 'date', 'time', 'value', 'error'], skipinitialspace=True)


```

```python
df_hr = {} #high temporal resolution dataset
```

```python
for year in (range(2000, 2024)):
    file_path = f'{env_data_dr}/BODC_tidal/{year}PTM.txt'  
    df_hr[year] = pd.read_csv(file_path, skiprows=range(0,11), delim_whitespace=True, header=None,names=['index', 'date', 'time', 'value', 'error'], skipinitialspace=True)
```

```python
df_hr[2021].head()
```

```python
# Concatenate the datasets from 2001 to 2024
df_hr_to_conc = []

for year in (range(2000, 2024)):
    file_path = f'{env_data_dr}/BODC_tidal/{year}PTM.txt' 
    df_hr_to_conc.append(pd.read_csv(file_path, skiprows=range(0,11), \
                                     delim_whitespace=True, header=None,\
                                     names=['index', 'date', 'time', 'value', 'error'], skipinitialspace=True))
```

```python
df_tide = pd.concat(df_hr_to_conc, ignore_index=True) 
```

```python
df_tide
```

```python
plt.plot(df_tide["date"][::200], df_tide["value"][::200])
plt.xlabel("time")
plt.ylabel("water surface height [m]")
plt.show()
```

### We need to clean up the dataset

```python
df_tide['date_hours'] = pd.to_datetime(df_tide['date'] + ' ' + df_tide['time'])
```

```python
# Set 'date_hours' as the index
df_tide.set_index('date_hours', inplace=True)
```

```python
df_tide = df_tide.drop(columns=["index", "date","time"])
```

```python
set(np.unique(df_tide["value"]))
```

### We remove data that are flagged (strings in the values)

```python
df_tide = df_tide[["value", "error"]].apply(pd.to_numeric, errors='coerce')
```

```python
plt.plot(df_tide.index, df_tide["value"])
plt.xlabel("time")
plt.ylabel("water surface height [m]")
plt.show()
```

We can now downsample the time interval, for example, with a 1h interval

```python
df_tide_1h = df_tide.resample("1H").mean() #resample in 24 h interval
```

```python
plt.plot(df_tide_1h.index, df_tide_1h["value"])
plt.ylabel("Surface elevation [m]")
plt.xlabel("time")

plt.show()
```

# Join the dataset together

```python
df_tide_1h.index
```

```python
df_tide_1h = df_tide_1h[(df_tide_1h.index.year>=2001)]
```

```python
ds_weather = ds_weather[(ds_weather.index.year<=2023)].drop(["YEAR", "MO", "DY", "HR"], axis=1)
```

```python
ds_env = ds_weather.join(df_tide_1h, how='outer')
```

```python
ds_env = ds_env.drop("date", axis=1)
```

## Quick explorations of the correlation in the dataset

```python
corrmat = ds_env.astype(float).corr().dropna(how="all", axis="columns").dropna(how="all", axis="rows")
f, ax = plt.subplots( figsize=(15, 15))
fig = sns.heatmap(corrmat, vmax=1, square=True, cmap="RdBu",  annot=True, vmin=-1);

plt.show()
```
